# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def unusual(a, b):
    master_lst = []
    for i in range(len(a)):
        for j in range(len(b)):
            if a[i] == b[j]:
                if a.count(a[i]) == 1:
                    if b.count(b[j]) == 1:
                        master_lst.append(a[i])
    if master_lst:
        print(master_lst)
    else:
        print('Уникальных чисел нет')

a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

unusual(a, b)
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
from time import *


def new_pet(pets, which_pet, pet_name, pet_age):
    dct = {}
    pack = [which_pet, pet_name, pet_age]
    dct['животное'] = pack
    dct['сытость'] = 100
    dct['энергия'] = 100
    dct['радость'] = 100
    pets.append(dct)
    return pets


def show_atributes(pets, pet_name):
    for pet in pets:
        for k, v in pet.items():
            if k == 'животное':
                if v[1] == pet_name:
                    print(f"сытость: {pet['сытость']}")
                    print(f"энергия: {pet['энергия']}")
                    print(f"радость: {pet['радость']}")


def feed_pet(pets, pet_name):
    for pet in pets:
        for k, v in pet.items():
            if k == 'животное' and v[1] == pet_name:
                if pet['сытость'] + 20 <= 100:
                    pet['сытость'] += 20
                else:
                    pet['сытость'] = 100

    return pets


def play(pets, pet_name):
    for pet in pets:
        for k, v in pet.items():
            if k == 'животное' and v[1] == pet_name:
                if pet['радость'] + 20 <= 100:
                    pet['радость'] += 20
                else:
                    pet['радость'] = 100

    return pets


def sleep(pets, pet_name):
    for pet in pets:
        for k, v in pet.items():
            if k == 'животное' and v[1] == pet_name:
                if pet['энергия'] + 20 <= 100:
                    pet['энергия'] += 20
                else:
                    pet['энергия'] = 100

    return pets


def debuff(pets, what_u_want, pet_name):
    for pet in pets:
        for k, v in pet.items():
            if k == 'животное' and v[1] == pet_name:
                if what_u_want == 1:
                    if pet['радость'] - 25 >= 0:
                        pet['радость'] -= 25
                    else:
                        pet['радость'] = 0
                if what_u_want == 2:
                    if pet['энергия'] - 25 >= 0:
                        pet['энергия'] -= 25
                    else:
                        pet['энергия'] = 0
                if what_u_want == 3:
                    if pet['сытость'] - 25 >= 0:
                        pet['сытость'] -= 25
                    else:
                        pet['сытость'] = 0

    return pets

def main():
    pets = []
    print('СПИСОК КОМАНД')
    print('1-завести питомца')
    print('2-поиграть/уложить спать/покормить')
    print('3-отобразить состояние питомца')
    print('4-stop')
    print()

    while True:

        main_answer = int(input('Ваш запрос: '))

        if main_answer == 1:
            which_pet = input('Какое животное вы хотите завести? ')
            pet_name = input('Какое имя вы хотите дать животному? ')
            pet_age = input('Возраст питомца? (лет): ')

            pets = new_pet(pets, which_pet, pet_name, pet_age)
            print()
            print(pets)
            print()

        elif main_answer == 2:
            print()
            pet_name = input('Имя животного: ')
            what_u_want = int(input('Что вы хотите сделать с животным?\n1-покормить\n2-поиграть\n3-уложить спать\n'))
            if what_u_want == 1:
                pets = feed_pet(pets, pet_name)
                pets = debuff(pets, what_u_want, pet_name)
            elif what_u_want == 2:
                pets = play(pets, pet_name)
                pets = debuff(pets, what_u_want, pet_name)
            elif what_u_want == 3:
                pets = sleep(pets, pet_name)
                pets = debuff(pets, what_u_want, pet_name)
            else:
                print('Такой команды нет')
            print(pets)

        elif main_answer == 3:
            print()
            pet_name = input('Имя животного: ')
            show_atributes(pets, pet_name)

        elif main_answer == 4:
            print()
            print('До свидания')
            break

        else:
            print('Такой команды нет')

main()
````

# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests

class Pokemons:
    def __init__(self):
        self.team = []

    def add_pokemon(self, name):
        if name in self.team:
            print(f'покемон {name} уже в команде')
            return
        response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{name}')
        if response.status_code == 200:
            pokemon_data = response.json()
            self.team.append(pokemon_data['name'])
            print(f'покемон {pokemon_data["name"]} добавлен в команду')
        else:
            print('Такого покемона нет')

    def delete_pokemon(self, name):
        if name in self.team:
            self.team.remove(name)
            print(f'{name} удален из команды')
        else:
            print(f'покемон {name} не найден в команде')

    def pokemon_team(self):
        if not self.team:
            print('В вашей команде нет покемонов')
            return
        for pokemon in self.team:
            print(pokemon)

    def find_pokemon(self, name):
        if name in self.team:
            print(f'покемон {name} уже в команде')
        else:
            print(f'покемон {name} не найден в команде')

    def fight(self, pokemon1_name, pokemon2_name):
        if pokemon1_name not in self.team or pokemon2_name not in self.team:
            print('Оба покемона должны быть в команде для боя')
        else:
            print(f'Начинаем бой между {pokemon1_name} и {pokemon2_name}!')

pokemon_manager = Pokemons()

pokemon_manager.add_pokemon('pikachu')
pokemon_manager.add_pokemon('ditto')
pokemon_manager.add_pokemon('charizard')

pokemon_manager.pokemon_team()
pokemon_manager.delete_pokemon('ditto')
pokemon_manager.pokemon_team()
pokemon_manager.find_pokemon('pikachu')

pokemon_manager.fight('pikachu', 'charizard')
````
